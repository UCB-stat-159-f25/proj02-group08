```python
import os
import matplotlib as mpl
import requests

# Download Comic Neue Regular
url = "https://github.com/cymon/comic-neue/raw/master/fonts/ComicNeue-Regular.ttf"
font_path = "/tmp/ComicNeue-Regular.ttf"

with open(font_path, "wb") as f:
    f.write(requests.get(url).content)

# Install font into Matplotlib's font directory
mpl_font_dir = os.path.join(mpl.get_data_path(), "fonts", "ttf")
!cp /tmp/ComicNeue-Regular.ttf "{mpl_font_dir}"

print("Font copied to:", mpl_font_dir)
```
